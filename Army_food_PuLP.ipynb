```python
df = pd.read_excel("C:/ISYE6040/data/diet.xls", sheet_name="Sheet1")
df
```

```python
#extracting min nutrition requirements
min_nutrition = df.iloc[-2,3:].tolist()
```

```python
print(min_nutrition)
```

```python
#extracting max nutrition requirements 
max_nutrition = df.iloc[-1,3:].tolist()
print(max_nutrition)
```

```python
#nutrients
nutrients = df.columns[3:].tolist()
```

```python
df = df.iloc[:-3]
df
```

```python
#creating the problem
prob = plp.LpProblem("army_diet", plp.LpMinimize)
```

```python
#creating decision variable
food_vars = plp.LpVariable.dicts("Food", df["Foods"], lowBound=0, cat="Continuous")
```

```python
prob += plp.lpSum(df.loc[i, 'Price/ Serving'] * food_vars[df.loc[i,'Foods']] for i in df.index)
#prob
```

```python
#nutrition constrnts
for j, nutrient in enumerate(nutrients):
    prob += plp.lpSum([df.loc[i, nutrient] * food_vars[df.loc[i, 'Foods']] 
                    for i in df.index]) >= min_nutrition[j]
    prob += plp.lpSum([df.loc[i, nutrient] * food_vars[df.loc[i, 'Foods']] 
                    for i in df.index]) <= max_nutrition[j]
```

```python
#call solver
prob.solve()
```

```python
print("Status:", plp.LpStatus[prob.status])
print("Total Cost =", plp.value(prob.objective))
print("Selected Foods:")
for v in prob.variables():
    if v.varValue > 0:
        print(v.name, "=", v.varValue)
```

```python
#objective function and nutritional constraints are the same.
#we need to create binary variables to indicate wether a particular food has been selected
```

```python
#binary variables for food selection
food_selected = plp.LpVariable.dicts("Select", df['Foods'], cat='Binary')
food_selected
```

```python
#constraint for servings
for i in df.index:
    food = df.loc[i, 'Foods']
    prob += food_vars[food] >= 0.1 * food_selected[food]
    # maybe we whould also set an upperbound for serving size?
```

```python
#constraint for mutually excluding broccoli and celery in the same mixture.
prob += food_selected['Celery, Raw'] + food_selected['Frozen Broccoli'] <= 1
```

```python
#list of all possible sources of protein
meat_foods = ['Roasted Chicken', 'Poached Eggs', 'Scrambled Eggs', 
              'Bologna,Turkey', 'Frankfurter, Beef', 'Ham,Sliced,Extralean',
              'Kielbasa,Prk', 'Pork', 'Sardines in Oil', 'White Tuna in Water',
              'Chicknoodl Soup', 'Splt Pea&Hamsoup', 'Vegetbeef Soup',
              'Neweng Clamchwd', 'New E Clamchwd,W/Mlk', 'Crm Mshrm Soup,W/Mlk',
              'Beanbacn Soup,W/Watr']
```

```python
#constraint for selecting a minimum of 3 different proteins
prob += plp.lpSum([food_selected[food] for food in meat_foods if food in food_selected]) >= 3
```

```python
#calling solver
prob.solve()
```
